```python
import numpy as np
import matplotlib.pyplot as plt
import glob
plt.style.use('science.mplstyle')
```

# Conservative Love Number

```python
# data_dir_cons = "./n_eq_1/Conservative/"
data_dir_cons = "./Data/n_eq_N_eq_1/"
```

```python
files = glob.glob(data_dir_cons+"*")

files = np.sort(files)

data_vals = []
for file in files:
    data_bkg = np.loadtxt(file+"/bkg.dat")
    data_love =  np.loadtxt(file+"/vals.dat")
    data_vals.append(np.array([data_bkg,data_love], dtype=object))
    
data_vals = np.array(data_vals, dtype=object)
```

```python
k2vals = np.array([2.4198937486*1e-1, 6.8656738911*1e-2, 3.2438798694*1e-2, 2.3293063321*1e-2,
                  1.7360105151*1e-2]) #Adiabatic values obtained from Binnigton and Poisson
lines = ["-","--","-.",(5, (10, 3)),(0, (1, 1))]
```

```python
form_names = ["omega","k2","normk2", "hat_tau_bulk", "p2_bulk", "norm_bulk", 
              "hat_tau_shear", "p2_shear", "norm_shear"]
```

## Figure 1

```python
fig, ax = plt.subplots(1,2, figsize = (12,4))
for j in range(len(k2vals)):
    data = data_vals[j][1]
    k2 = k2vals[j]
    Ca = data_vals[j][0][-1]
    ax[0].plot(data[:,0],np.abs(1-data[:,1]/k2)*100,label = r'$C={:0.2f}$'.format(Ca),linestyle = lines[j])
    
    data = np.abs(data_vals[j][1])
    k2 = k2vals[j]
    Ca = data_vals[j][0][-1]
    ax[1].plot(data[:,0],np.abs(data[:,1]), label = r'$C={:0.2f}$'.format(Ca),linestyle = lines[j])
ax[0].set_xlabel(r'$\Omega$')
ax[0].set_ylabel(r'Percent Difference')
ax[0].set_xlim(3e-3,4e-2)
ax[0].set_ylim(1e-4,2)
ax[0].set_yscale('log')

ax[1].set_xlim(0.3,0.5)
ax[1].set_ylim(1e-4,1e3)
ax[1].set_yscale('log')
ax[1].set_xlabel(r'$\Omega$')
ax[1].set_ylabel(r'$\left|K_2(\Omega)\right|$')
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, bbox_to_anchor = (0.4,0.5) )
plt.savefig('./Figures/conservative_tides.pdf')
```

## Viscosity Figure

```python
fig, ax = plt.subplots(1,2, figsize = (12,4),sharex=True)
for j in range(len(k2vals)):
    data = data_vals[j][1]
    k2 = k2vals[j]
    Ca = data_vals[j][0][-1]
    ax[0].plot(data[:,0],np.abs(data[:,4]),label = r'$C={:0.2f}$'.format(Ca),linestyle = lines[j])
    ax[1].plot(data[:,0],np.abs(data[:,7]),label = r'$C={:0.2f}$'.format(Ca),linestyle = lines[j])
    
ax[0].set_yscale('log')
ax[0].set_xlim(3e-3,4e-2)
ax[0].set_ylim(0.5*1e-3,5e-2)
ax[1].set_ylim(2,15)
ax[0].set_xlabel(r'$\Omega$')
ax[1].set_xlabel(r'$\Omega$')
ax[0].set_ylabel(r'$p_{2,\mathrm{bulk}}$')
ax[1].set_ylabel(r'$p_{2,\mathrm{shear}}$')
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, bbox_to_anchor = (0.85,0.87) ,ncol=2,handleheight=2.4, labelspacing=0.05)
plt.savefig('./Figures/p2_vals.pdf')
```
